```python
import pandas as pd
```

```python
data = pd.read_csv('/content/House Price India.csv.zip')
```

```python
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
stats = data["Price"].describe().reset_index()
stats["Price"] = round(stats["Price"],2)
```

```python
stats
```

```python
data.isna().sum().sum()
```

```python
data.duplicated().sum()
```

```python
# for na values
data.dropna(inplace = True)
# for duplicated values
data.drop_duplicates(inplace= True)
```

```python
data.columns
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
data.groupby("condition of the house")["Price"].mean().sort_values(ascending=False).plot(kind = "bar")
plt.title("Condition of the house vs Price")
plt.ylabel("Mean Price")
plt.xlabel("Condition of the house")
plt.show()
```

```python
data.columns
```

```python
data[["number of bedrooms", "number of bathrooms"]]
```

```python
X = data[['number of bedrooms', 'number of bathrooms', 'living area', 'condition of the house', 'Number of schools nearby']]
y = data[["Price"]]
```

```python
X
```

```python
y
```

```python
data.shape[0] * 0.8
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2)
```

```python
X_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid = {
    "criterion" : ["mse", "friedman_mse", "nae"],
    "splitter" : ["best", "random"],
    "max_depth" : [None, 10, 20, 30, 40, 50],
    "min_samples_split" : [2,5,10],
    "min_samples_leaf" : [1,2,4]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = tree_model, param_grid = param_grid)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predslr = lr.predict(X_test)
```

```python
mean_absolute_error(y_test, predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor()
```

```python
param_gridfr = {
    "max_depth" : [5,10,15],
    "n_estimators" : [2,3,4,5,6,7,8,9,10]
}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridfr)
```

```python
gridrfr.fit(X_train, y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
rfrpredictions = gridrfr.predict(X_test)
```

```python
mean_absolute_error(y_test,rfrpredictions)
```

```python
gridrfr
```

```python
import joblib
joblib.dump(gridrfr,'model.pkl')
```

```python
%%writefile app.py
import streamlit as st
import joblib
import numpy as np

st.title("Rizan's House Price Prediction")

st.divider()

st.write("House Prediction App")

st.divider()

st.write("This app predicts the price of a house based on its features.")

st.divider()

bedrooms = st.number_input("Number of bedrooms", min_value=0, max_value=10)
bathrooms = st.number_input("Number of bathrooms", min_value=0.0, max_value=8.0, format="%.2f")
livingarea = st.number_input("Living area", min_value=0, max_value=5000)
condition = st.number_input("Condition of the house", min_value=0, max_value=5)
numberofschools = st.number_input("Number of schools", min_value=0, max_value=5)

st.divider()

X = [[bedrooms, bathrooms, livingarea, condition, numberofschools]]

predictbutton = st.button("Predict!")

if predictbutton:

  st.balloons()

  X_array = np.array(X)
  model = joblib.load("model.pkl")

  prediction = model.predict(X_array)

  st.write(f"The predicted price of the house is {prediction[0]:,.2f}")
```
